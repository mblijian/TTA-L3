```python
def maze_escape():
    # maze escape with while condition loop and the variable move_forward.
    import time
    move_forward = 0
    Laser1 = got.read_distance_data(21)

    while move_forward< 7:
        Laser1 = got.read_distance_data(21)
        if Laser1 >15:
            got.mecanum_move_speed(0,20)
        else:
            got.mecanum_stop()
            move_forward+=1
            print(f"This is move forward: {move_forward}.")
            time.sleep(1)
            got.mecanum_turn_speed_times(3,30,90,2)
            
            Laser1 = got.read_distance_data(21)
            if Laser1<15:
                got.mecanum_turn_speed_times(3,30,180,2)
                time.sleep(0.5)       
```

## Session 8/9

1. function library

```python
def pick_up():
    import time
    got.mechanical_clamp_release()
    time.sleep(2)
    got.mechanical_joint_control(0,30,30,1000)
    time.sleep(1)
    got.mechanical_joint_control(0,0,-70,1000)
    time.sleep(1)
    got.mechanical_clamp_close()
    time.sleep(2)
    got.mechanical_joint_control(0,30,30,1000)
    time.sleep(1)
    
def put_down():
    import time
    got.mechanical_joint_control(-90,30,30,1000)
    time.sleep(1)
    got.mechanical_joint_control(-90,0,-80,1000)
    time.sleep(1)
    got.mechanical_clamp_release()
    time.sleep(2)
    got.mechanical_joint_control(-90,30,30,1000)
    time.sleep(1)
    got.mechanical_joint_control(0,30,30,1000)
    time.sleep(1)
    got.mechanical_clamp_close()
    time.sleep(2)

def move_stop():
    while True:
        Laser1 = got.read_distance_data(21)
        if Laser1 > 14:
            got.mecanum_move_speed(0,15)
        else:
            got.mecanum_stop()
            time.sleep(0.5)
            break

def pick_up_100():
    import time
    got.mechanical_joint_control(0,30,30,1000)
    time.sleep(1)
    move_stop()
    pick_up()

    Laser1 = got.read_distance_data(21)
    while Laser1<12:
        Laser1 = got.read_distance_data(21)
        if Laser1<4:
            got.mecanum_move_speed_times(1,5,6,1)
            time.sleep(1)
            pick_up()
            time.sleep(2)

    print("I have picked up the token.")

def maze_escape():
    # maze escape with while condition loop and the variable move_forward.
    import time
    move_forward = 0
    Laser1 = got.read_distance_data(21)

    while move_forward< 6:
        Laser1 = got.read_distance_data(21)
        if Laser1 >15:
            got.mecanum_move_speed(0,20)
        else:
            got.mecanum_stop()
            move_forward+=1
            print(f"This is move forward: {move_forward}.")
            time.sleep(1)
            got.mecanum_turn_speed_times(3,30,90,2)
            
            Laser1 = got.read_distance_data(21)
            if Laser1<15:
                got.mecanum_turn_speed_times(3,30,180,2)
                time.sleep(0.5)    

def Coley_recognition():
    import time
    got.load_models(["face_recognition"])
    face_info = got.get_face_recognition_total_info()

    while True:
        face_info = got.get_face_recognition_total_info()
        if not face_info:
            got.mecanum_move_speed(0,15)
        else:
            got.mecanum_stop()
            break
got.play_audio_tts("Hello Coley, here's your package, have a nice day!!",0,False)   
```
