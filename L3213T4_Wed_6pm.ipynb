2. function library

```python
# 1. pick up funciton
def pick_up():
    import time
    got.mechanical_joint_control(0,30,30,1000)
    time.sleep(2)
    got.mechanical_clamp_release()
    time.sleep(2)
    got.mechanical_joint_control(0,0,-70,1000)
    time.sleep(2)
    got.mechanical_clamp_close()
    time.sleep(2)
    got.mechanical_joint_control(0,30,30,1000)
    time.sleep(2)

# 2. put down function
def put_down():
    import time
    got.mechanical_joint_control(90,30,30,1000)
    time.sleep(2)
    got.mechanical_joint_control(90,0,-80,1000)
    time.sleep(2)
    got.mechanical_clamp_release()
    time.sleep(2)
    got.mechanical_joint_control(90,30,30,1000)
    time.sleep(2)
    got.mechanical_joint_control(0,30,30,1000)
    time.sleep(2)
    got.mechanical_clamp_close()
    time.sleep(2)

# 3. move and stop
def move_stop():
    while True:
        Laser1 = got.read_distance_data(21)
        if Laser1> 10:
            got.mecanum_move_speed(0,20)
        else:
            got.mecanum_stop()
            break

# 4. maze escape
def maze_escape():
    import time
    move_forward = 0
    while move_forward<8:
        Laser1 = got.read_distance_data(21)
        if Laser1> 10:
            got.mecanum_move_speed(0,20)
        else:
            got.mecanum_stop()
            time.sleep(0.5)
            got.mecanum_turn_speed_times(3,30,90,2)
            move_forward+=1
            print(f"This is move forward:{move_forward}.")
            
            Laser1 = got.read_distance_data(21)
            if Laser1<10:
                got.mecanum_turn_speed_times(2,30,180,2)
                time.sleep(0.5)

# 5. pick up 100% function
def pick_up_guarantee():
    import time
    move_stop()
    pick_up()
    time.sleep(1)

    Laser1 = got.read_distance_data(21)
    while Laser1<12:
        Laser1 = got.read_distance_data(21)
        if Laser1<5:
            got.mecanum_move_speed_times(1,5,5,1)
            time.sleep(0.5)
            pick_up()
            time.sleep(2)

    got.show_light_rgb_effect(0,255,0,3)
    got.play_audio_tts("I got the token! Let's go through the maze now!", 0, False)
```

```python
# final mission
pick_up_guarantee()
time.sleep(0.5)
maze_escape()
time.sleep(0.5)
put_down()
```
