## Session 9/10

```python
# 1. maze escape function
def maze_escape():
    import time
    move_forward = 0
    while move_forward<4:
        Laser1 = got.read_distance_data(21)
        if Laser1> 10:
            got.mecanum_move_speed(0,20)
        else:
            got.mecanum_stop()
            time.sleep(0.5)
            got.mecanum_turn_speed_times(3,30,90,2)
            move_forward+=1
            print(f"This is move forward:{move_forward}.")
            
            Laser1 = got.read_distance_data(21)
            if Laser1<10:
                got.mecanum_turn_speed_times(2,30,180,2)
                time.sleep(0.5)

# 2. pick up function
def pick_up():
    import time
    got.mechanical_clamp_release()
    time.sleep(2)
    got.mechanical_joint_control(0,30,30, 1000)
    time.sleep(1)
    got.mechanical_joint_control(0,0,-70, 1000)
    time.sleep(1)
    got.mechanical_clamp_close()
    time.sleep(2)
    got.mechanical_joint_control(0,30,30, 1000)
    time.sleep(1)

# 3. put down function
def put_down():
    import time
    got.mechanical_joint_control(90,30,30, 1000)
    time.sleep(1)
    got.mechanical_joint_control(90,0,-80, 1000)
    time.sleep(1)
    got.mechanical_clamp_release()
    time.sleep(2)
    got.mechanical_joint_control(90,30,30, 1000)
    time.sleep(1)
    got.mechanical_joint_control(0,30,30, 1000)
    time.sleep(1)
    got.mechanical_clamp_close()
    time.sleep(2)

# 4. face recognition function
def face_recognition():
    while True:
        face_info = got.get_face_recognition_total_info()
        if not face_info:
            got.mecanum_move_speed(0,20)
        else:
            got.mecanum_stop()
            break
```
